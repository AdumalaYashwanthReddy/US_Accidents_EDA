### Start Time

```python
df.Start_Time
```

```python
df['Start_Time'] = df['Start_Time'].astype(str)  # Convert to string
df['Start_Time'] = df['Start_Time'].str.split('.').str[0]  # Remove anything after the dot
df['Start_Time'] = pd.to_datetime(df['Start_Time'])  # Convert to datetime
```

- Figure out how to show percentage.

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

- High no.of accidents occur between 6am and 9 am (probably in a hurry to et to work).
- Next highest percentage is 3pm to 6pm.

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

- The distribution of accidents by hour is not same on weekends as on weekdays.

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

- On sundays the peak occurs between 10am and 3pm, unlike weekdays.

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

- The no.of accidents are higher at the Start of the year and, a little bit higher during the End of the year, and low during the summer.

### Weather

```python
weather_accidents = df['Weather_Condition'].value_counts()

print(weather_accidents.head(10))
```

```python
plt.figure(figsize=(12, 6))
weather_accidents.head(10).plot(kind='bar', color='lightblue')
plt.title('Top 10 Weather Conditions by Accident Count')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
df_cleaned = df[df['Weather_Condition'].notnull() & (df['Weather_Condition'] != 'Unknown')]

weather_accidents_cleaned = df_cleaned['Weather_Condition'].value_counts()
```

```python
weather_mapping = {
    'Clear': 'Clear',
    'Partly Cloudy': 'Cloudy',
    'Mostly Cloudy': 'Cloudy',
    'Overcast': 'Cloudy',
    'Light Rain': 'Rain',
    'Heavy Rain': 'Rain',
    'Rain': 'Rain',
    'Snow': 'Snow',
    'Fog': 'Fog',
}

df['Weather_Category'] = df['Weather_Condition'].map(weather_mapping)

weather_category_counts = df['Weather_Category'].value_counts()

# Visualize the accident counts by grouped weather category
plt.figure(figsize=(10, 6))
weather_category_counts.plot(kind='bar', color='lightcoral')
plt.title('Accidents by Grouped Weather Categories')
plt.xlabel('Weather Condition')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```
